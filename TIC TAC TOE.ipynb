```python
def printer(a):
    print(f' {a[0]} | {a[1]} | {a[2]}')
    print(f'--- --- --- ')
    print(f' {a[3]} | {a[4]} | {a[5]}')
    print(f'--- --- --- ')
    print(f' {a[6]} | {a[7]} | {a[8]}')
```
